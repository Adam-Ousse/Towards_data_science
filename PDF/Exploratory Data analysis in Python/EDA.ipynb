```python
import pandas as pd

# Sample DataFrame
data = {
    'text': [
        "cat", "dog", "elephant", "bird", "fish123", 
        "apple", "pear", "orange", "grape", "22 kiwi",
        "The quick brown fox", "Jumped over", "31 lazy dogs"
    ]
}
df = pd.DataFrame(data)

# 1. Matches any character followed by 'a'
print(".a")
print(df[df['text'].str.contains("^.a", regex=True)])

# 2. Matches any string with three consecutive digits*
print("digits")
print(df[df['text'].str.contains("^\\d{2}")])

# 3. Matches any string with one or more alphanumeric characters
print("word")
print(df[df['text'].str.contains("\\w+")])

# 4. Matches any string with a whitespace character
print("whitespace")
print(df[df['text'].str.contains("\\s")])

# 5. Matches any string containing a vowel
print("vowel")
print(df[df['text'].str.contains("^[aeiou]")])

print("novowel")
# 6. Matches any string without a vowel
print(df[df['text'].str.contains("^[^aeiou]")])

# 7. Matches any string that ends with "dog"
print("enddog")
print(df[df['text'].str.contains("dog$")])
```
