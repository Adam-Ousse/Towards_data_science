```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np
```

```python
d = norm.rvs(5,2,1000)
w= norm.rvs(5,2,1000)
# sns.scatterplot(x=d, y=w)
sns.histplot(d)
sns.histplot(w)

plt.show()
```

```python
df = sns.load_dataset("tips")
# hue_colors= {"Yes":"black","No":"red"}
hue_colors= {"Yes":"#FF8343","No":"#179BAE"}
sns.scatterplot(x="total_bill",y="tip",data=df,hue="smoker",hue_order=["No","Yes"],palette=hue_colors,size="size")
plt.show()
```

```python
sns.countplot(x="smoker",data=df,hue="sex",palette={"Male":"#179BAE","Female":"#FF8343"})
plt.show()
```

```python

sns.relplot(x="total_bill",y="tip",data=df,kind="scatter",row="time",size="size",hue="size",style="smoker")
plt.show()
```
